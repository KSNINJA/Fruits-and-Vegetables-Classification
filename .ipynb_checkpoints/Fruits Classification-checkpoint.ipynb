```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import PIL
import PIL.Image
import pathlib




```

```python
train_valid_dataset_dir = pathlib.Path("./fruits-360/Training")
fruit1 = "Apple Golden 1"
fruit1_list = list(train_valid_dataset_dir.glob(fruit1 + "/*"))
PIL.Image.open(fruit1_list[0])
```

```python
img_height = 100
img_width = 100
batch_size = 500

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    train_valid_dataset_dir,
    validation_split=0.2,
    subset = 'training',
    seed = 123,
    image_size = (img_height, img_width),
    batch_size = batch_size
)
train_ds.__dict__


val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  train_valid_dataset_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3,3, i+1)
    plt.imshow(PIL.Image.open(pathlib.Path(train_ds.file_paths[i])))
    plt.axis("off")
plt.show()
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

num_classes = 131

model = tf.keras.Sequential([
  tf.keras.layers.experimental.preprocessing.Rescaling(1/255),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(num_classes)
])


model.compile(optimizer="adam", loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
h = model.fit(train_ds, validation_data=val_ds, epochs = 10)
```
